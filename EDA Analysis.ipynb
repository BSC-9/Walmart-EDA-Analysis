<font size="6">Finding Central Tendencies</font>

```python
#Store

print('Max: ',walnew.max()['Store'])
print('\nMode: ',walnew.mode()['Store'])
print('\nMean: ',walnew['Store'].mean())
print('\nMedian: ',walnew['Store'].median())
```

```python
#Weekly_Sales

print('Max: ',walnew.max()['Weekly_Sales'])
print('\nMode: ',walnew.mode()['Weekly_Sales'])
print('\nMean: ',walnew['Weekly_Sales'].mean())
print('\nMedian: ',walnew['Weekly_Sales'].median())
```

```python
#Holiday_Flag

print('Max: ',walnew.max()['Holiday_Flag'])
print('\nMode: ',walnew.mode()['Holiday_Flag'])
print('\nMean: ',walnew['Holiday_Flag'].mean())
print('\nMedian: ',walnew['Holiday_Flag'].median())
```

```python
#Unemployment

print('Max: ',walnew.max()['Unemployment'])
print('\nMode: ',walnew.mode()['Unemployment'])
print('\nMean: ',walnew['Unemployment'].mean())
print('\nMedian: ',walnew['Unemployment'].median())
```

<font size="6">Data visualization</font>

```python
plt.style.use('ggplot')
```

```python
#BAR GRAPH


x = walnew["Store"]
y = walnew["Unemployment"]

plt.xlabel("Store", fontsize=10)
plt.ylabel("Unemployment", fontsize=10)
plt.title("Store-wise Unemployment")

plt.bar(x, y)

plt.show()
```

```python
#HISTOGRAM

x = walnew['Weekly_Sales']

plt.xlabel('Weekly_Sales', fontsize=10)

plt.hist(x, bins='auto', histtype='bar', rwidth=0.8)

plt.show()
```

```python
#SCATTER PLOT


x = walnew["Store"]
y = walnew["Weekly_Sales"]

plt.xlabel("Store", fontsize=10)
plt.ylabel("Weekly Sales", fontsize=10)
plt.title("Weekly Sales by Store")

plt.scatter(x, y)

plt.show()
```
